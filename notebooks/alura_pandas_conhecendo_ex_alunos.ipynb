```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/desafios/alunos.csv'
dados = pd.read_csv(url)
dados = dados.fillna(0)
```

```python
dados
```

```python
dados['Pontos_extras'] = dados['Notas'] * 0.4
dados.head()
```

```python
dados['notas_finais'] = dados['Pontos_extras'] + dados['Notas']
dados['notas_finais'] = dados['notas_finais'].apply(lambda x: 10.0 if x > 10.0 else x)
dados.head()
```

```python
dados
```

```python
dados['Aprovado_final'] = dados['notas_finais'].apply(lambda x: True if x >= 6 else False)
dados
```

```python
dados_reprovados_antes_aprovados_final = dados.query('Aprovado == False & Aprovado_final == True')
dados_reprovados_antes_aprovados_final
```
